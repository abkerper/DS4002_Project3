```python
import os
from PIL import Image
import numpy as np

# Define folder path and target size
folder_path = "/content/DS4002_Project3/DATA"
output_file = "dataset.npy"
target_size = (224, 224)  # Resize to 224x224

# Load and preprocess images
image_data = []
for file_name in os.listdir(folder_path):
    if file_name.endswith(('.jpg', '.png', '.jpeg')):  # Check for image files
        img_path = os.path.join(folder_path, file_name)
        img = Image.open(img_path).resize(target_size)  # Resize image
        img_array = np.array(img)  # Convert to numpy array
        image_data.append(img_array)

# Convert to numpy array and save
image_data = np.array(image_data)
np.save(output_file, image_data)

print(f"Dataset saved as {output_file} with shape: {image_data.shape}")
```
